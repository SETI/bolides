## 1. Plot the light curve of a bolide detected by GLM

```python
from bolides import Bolide
b = Bolide("5dcf3a936f9c273b38592eae")
lc = b.to_lightcurve().normalize()
lc.plot(marker='o');
```

## 2. Extract light curves of the same event from YouTube videos

```python
from bolides import youtube_photometry
ytlc1 = youtube_photometry("cH1Uz06WVHs").normalize()
ytlc2 = youtube_photometry("kWUlykY9YBw").normalize()
ytlc3 = youtube_photometry("YbQ0RoxFezQ").normalize()
```

## 3. Plot all light curves together!

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")

fig = plt.figure(figsize=(6, 8), dpi=200)

ax = plt.subplot(411)
lc[2:-6].plot(ax=ax, marker='o', label="GLM")

ax = plt.subplot(412)
ytlc1[20:-16].plot(ax=ax, marker='o', label="YouTube 1");

ax = plt.subplot(413)
ytlc2[85:-13].plot(ax=ax, marker='o', label="YouTube 2");

ax = plt.subplot(414)
ytlc3[50:-264].plot(ax=ax, marker='o', label="YouTube 3");
```
